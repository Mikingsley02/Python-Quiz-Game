```python
#10_Questions quiz game

print ("WELCOME!")

start = input("Are you ready to play? ")
if start.lower() != "yes":
    quit()   
else:
    print ("Cool! let's play :) ")
    score = 0
   
answer = input("Which country was the first to lift a ban on the use of face mask after the world's most deadly pandemic in 2019? ")
if answer.lower() == "united kingdom":
    print ("Correct!")
    score = score + 1
else:
    print ("Incorrect!")


answer = input("What is the full meaning of CPU? ")
if answer.lower() == "central processing unit":
    print ("Correct! you must be a computer wizard")
    score += 1
else:
    print ("Incorrect!")
    

answer = input("After the last supper, Jesus was layed-off, Peter was pissed-off, who was payed off? ")
if answer.lower() == "judas":
    print ("Wow! thats correct, even Jesus would be pleased :)")
    score = score + 1
else:
    print ("Incorrect!, go back to sunday school classes")
    

answer = input("If i give you a basket full of 5 oranges, 3 pinapples, 4 onions, 7 Bananas and I take away 3 oranges. How many fruits would be left? ")
if answer == "12":
    print ("That's correct, you're a genus")
    score = score + 1
else:
    print ("Sorry, you failed. Onion is not a fruit")
    

answer = input("What is the full meaning of NYPD? ")
if answer.lower() == "new york police department":
    print ("Correct!")
    score = score + 1
else:
    print ("Sorry, that was incorrect.")
    

answer = input("In what year did Nigeria become a republic? ")
if answer == "1963":
    print ("Correct")
    score += 1 
else:
    print ("Incorrect")
    
    
answer = input("There are 11 players in a football team, how many players are in a football field excluding the goalkeepers? ")
if answer == "20":
    print ("Correct")
    score = score + 1
else:
    print ("Incorrect")
    
answer = input("What is the administrative capital of South Africa? ")
if answer.lower() == "pretoria":
    print ("Correct!")
    score = score + 1
else:
    print ("Incorrect")
    
    
print ("Your total score is: " + str(score))
print ("Your score percentage is: " + str((score/8)*100) + "%")

if score >= int("4"):
    print ("Congratulations, You Won!")
else:
    score <= int("3")
    print ("You Lost, Try Again!")
```
